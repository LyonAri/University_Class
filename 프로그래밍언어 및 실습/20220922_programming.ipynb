### for문 복습

```python
for i in range (1, 10):
  for j in range (1, 10):
    print(f"{j:d}x{i:d}={i*j:2d}", end=" ")
  print(end="\n") 
```

```python
for i in range(5):
  for j in range(i+1, 5):
    print(" ", end="")
  for j in range(2*i+1):
    print("*", end="")
  print(end = "\n")

for k in range(4, -1, -1):
  for l in range(k-1, 4):
    print(" ", end="")
  for l in range(2*(k-1)+1):
    print("*", end="")
  print(end = "\n")
```

### While문

```python
i = 0
while i<=10:
  print(i, end=" ")
  i += 1
else:
  print("done")
print("Hello World!")
```

```python
i = 0
while i<=10:
  if(i%2) == 0:
    print(i, end=" ")
  i += 1
```

```python
i = 0
while i<=10:
  if(i%2) == 1:
    i += 1
    continue
  print(i, end=" ")
  i += 1
```

```python
i, res = 0, 0
while True:
  if i > 10:
    break
  res += i
  i += 1
print(res)
```

```python
while True:
  n = int(input("단 입력(1~9) [종료 0]: "))
  if n>=10 or n < 0:
    print("1~9사이의 숫자를 입력하세요.")
    continue
  elif n == 0:
    print("프로그램을 종료합니다.")
    break

  for i in range(1, 10):
    print(f"{n:d}x{i:d}={n*i:2d}")
```

### while, break, continue 연습문제 (14~17)

```python
# for문

n = int(input())
sum = 0
for i in range(1, n+1):
  if i%3 == 0:
    sum = sum + i

print(sum)
```

```python
# while문

n = int(input())
sum = 0
i=0

while i <= n:
  if i%3 == 0:
    sum = sum + i
  i=i+1
print(sum)
```

```python
# for문

sum = 0

for i in range(1, 1001):
  sum = sum+i

print(sum)
```

```python
# while문

sum = 0
i = 0

while i <= 1000:
  sum = sum+i
  i=i+1
print(sum)
```

```python
n = int(input())

for i in range(2, n):
  if n % i == 0:
    print("Not Prime")
    break
  else:
    print("Prime")
    break
```

```python
def my_sum(n):
  """이 함수는 0부터 입력한 n까지 더하는 프로그램입니다.
* n (Argument) : 입력한 n까지 더함
* return : 계산된 결과를 반환함 """
  res = 0
  for i in range(n+1):
    res += i
  return res

help(my_sum)
print(my_sum(10))
```

### 함수 연습문제 

```python
def c_f(c):
  return (c*1.8) + 32.0

n=float(input())

print(c_f(n))
```

```python
def bmi(h, w):
  return w / (0.01*h)**2

h, w = input().split()
w=int(w)
h=int(h)

print("%.2f" % bmi(h, w))
```

```python
def time(t):
  h = t/3600
  m = (t-3600*h)/60
  s = t%60
  return h, m, s

t=int(input())
print(time(t))
```
