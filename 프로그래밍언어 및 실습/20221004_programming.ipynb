### 지역변수와 전역변수

```python
tmp = 10

def print_tmp1():
  tmp = "Hello"
  print(tmp)

def print_tmp2():
  tmp = 3.141592
  print(tmp)

print_tmp1()
print_tmp2()
print(tmp)
```

```python
tmp = 10

def print_tmp1():
  tmp = "Hello"
  print(tmp)

def print_tmp2():
  global tmp
  print(tmp)

print_tmp1()
print_tmp2()
print(tmp)
```

```python
def my_add(*args):
  res = 0
  for num in args:
    res += num
  return res

def my_sub(*args):
  for idx, num in enumerate(args):
    if idx == 0:
      res = num
      continue
    res -= num
  return res

def my_calc(op_func, *args):
  res = op_func(*args)
  return res

res = my_calc(my_add, 5, 3, 2, 2)
print(res)
res = my_calc(my_sub, 5, 3)
print(res)
```

### 람다식

```python
def add1(a, b):
  return a+b

print(add1(5, 3))
```

```python
add2 = lambda a, b : a+b
print(add2(5, 3))

print((lambda a, b : a + b)(5, 3))
```

```python
my_add = lambda x, y, z : x + y + z
my_sub = lambda x, y, z : x - y - z

def my_calc(my_func, a, b, c):
  return my_func(a, b, c)

print(my_calc(my_add, 12, 5, 3))
print(my_calc(my_sub, 12, 5, 3))
```

```python
my_add = lambda * args : sum(args)

print(my_add(12, 5))
print(my_add(12, 5, 3))
print(my_add(12, 5, 3, 1))
```

```python
my_sub = lambda *args : 2*args[0] - sum(args)

print(my_sub(12, 5))
print(my_sub(12, 5, 3))
print(my_sub(12, 5, 3, 1))
```

```python
def my_calc(func, x):
  return func(x)

square = lambda x : x**2
cube = lambda x : x**3
root = lambda x : x**0.5

print(my_calc(square, 4))
print(my_calc(cube, 4))
print(my_calc(root, 4))
```

### 연습문제 ( 5 ~ 8 )

```python
def is_prime_number(x):
  for i in range(2, x):
    if x % i == 0:
    	return False
  return True

x = int(input())
if is_prime_number(x)==False:
  print("Not Prime")
else:
  print("Prime")
```

```python
def factorial(n):
    if n == 1:
      return 1
    else:
      return n * factorial(n - 1)

def combination(n, r):
  return factorial(n)/(factorial(n-r)*factorial(r))

n, r = input().split()
n = int(n)
r = int(r)
print("%d" % combination(n, r))
```

```python
import math

def exp(x, n):
  e = 0
  for i in range(n+1):
    e += x**i / math.factorial(i)
  return e

x, n = input().split()
x=int(x)
n=int(n)

print("%.2f" % exp(x, n))
```

```python
def hanoi(n, from_pos, tmp, to_pos):
  if n==1:
    print(f"{1}: {from_pos} -> {to_pos}")
    return
  else:
    hanoi(n-1, from_pos, to_pos, tmp)
    print(f"{n}: {from_pos} -> {to_pos}")
    hanoi(n-1, tmp, from_pos, to_pos)

n = int(input())
hanoi(n, 'A', 'B', 'C')
```

### 연습문제 정답으로 공개된 코드

```python
def factorial(n):
  res = 1
  for i in range (2, n+1):
    res*=i
  return res

def combination(n, r):
  return factorial(n)/(factorial(n-r)*factorial(r))

n, r = input().split()
n = int(n)
r = int(r)
print(int(combination(n, r)))
```

```python
def factorial(n):
  res = 1
  for i in range (2, n+1):
    res*=i
  return res

def exp(x, n):
  e = 0
  for i in range(n+1):
    e += (lambda x:x**i)(x)/factorial(i)
  return e

x, n = input().split()
x=int(x)
n=int(n)

print("%.2f" % exp(x, n))
```
