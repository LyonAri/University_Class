### while과 break, continue 연습문제 (14 ~ 17)

```python
n = int(input())
sum = 0
for i in range(1, n+1):
  if i%3 == 0:
    sum = sum + i

print(sum)
```

```python
sum = 0

for i in range(1, 1001):
  sum = sum+i

print(sum)
```

```python
n = int(input())

for i in range(2, n):
  if n % i == 0:
    print("Not Prime")
    break
  else:
    print("Prime")
    break
```

```python
n = int(input())
count = 1
while n > 10:
  n = n/10
  count += 1

print(count)

# 입력한 수를 10으로 나누면서 자릿수 계산
```
