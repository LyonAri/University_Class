### 조건문 연습문제 (5 ~ 7)

```python
a, b, c = input().split()

a = int(a)
b = int(b)
c = int(c)

if a + b > c:
  if a + c > b:
    if b + c > a:
      print("삼각형 O")
else:
  print("삼각형 X")
```

```python
k, m, e, s = input().split()

k = int(k)
m = int(m)
e = int(e)
s = int(s)

avg = (k + m + e + s) / 4
print(avg, end = ", ")

if avg >= 90:
  print("장학금 O")
else:
  print("장학금 X")
```

```python
year = int(input())

if year % 4 == 0 and year % 100 != 0 or year % 400 == 0:
  print("윤년")
else:
  print("평년")
```
