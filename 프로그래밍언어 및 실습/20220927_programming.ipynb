### 함수 연습문제 (1 ~ 3)

```python
def c_f(c):
  return (c*1.8) + 32.0

n=float(input())

print("%.2f" % c_f(n))
```

```python
def bmi(h, w):
  return w / (0.01*h)**2

h, w = input().split()
w=int(w)
h=int(h)

print("%.2f" % bmi(h, w))
```

```python
def time(t):
  h = t // 3600
  m = (t - (h * 3600)) // 60
  s = t % 60
  print(h, m, s)

t=int(input())
time(t)
```

```python
def my_add(*args):
  res = 0
  for i in args:
    res += i
  return res

print(my_add(5, 3))
print(my_add(5, 3, 2))
print(my_add(5, 3, 2 ,1, 1, 1, 1, 1))
```

```python
def my_calc(op, *args):
  res = 0
  for idx, num in enumerate(args):
    if op == "+":
      if idx == 0:
        res = num
        continue

      res += num
    elif op == "-":
      if idx == 0:
        res = num
        continue
      res -= num
    else:
      print("[Error] Not Supported Operation")
      return
  return res

print(my_calc("+", 5, 2, 1, 1))
print(my_calc("-", 5, 2, 1))
```

```python
def print_n_times(n, *texts):
  for i in range(n):
    for text in texts:
      print(text, end = " ")
    print(end ="\n")

print_n_times(3, "안녕하세요.", "반갑습니다.", "홍길동입니다.")
```

```python
def print_n_times(*texts, n):
  for i in range(n):
    for text in texts:
      print(text, end = " ")
    print(end ="\n")

print_n_times("안녕하세요.", "반갑습니다.", "홍길동입니다.", 3)
```

### 함수 연습문제 (4 ~ )

```python
def fibo(n):
  if n <= 2:
    return 1
  return fibo(n-1)+fibo(n-2)

n = int(input())

print(fibo(n))
```

```python
def is_prime_number(x):
  for i in range(2, x):
    if x % i == 0:
    	return False
  return True

x = int(input())
if is_prime_number(x)==False:
  print("Not Prime")
else:
  print("Prime")
```

```python
def factorial(n):
    if n == 1:
      return 1
    else:
      return n * factorial(n - 1)

def combination(n, r):
  return factorial(n)/(factorial(n-r)*factorial(r))

n, r = input().split()
n = int(n)
r = int(r)
print("%d" % combination(n, r))
```

```python
def square_sum(n):
  if n == 1:
    return 1
  else:
    return n**2 + square_sum(n-1)

n = int(input())
print(square_sum(n))
```

```python
import math

def factorial(n):
    if n == 1:
      return 1
    else:
      return n * factorial(n - 1)

e_to_2 = 0
def exp(x, n)
  for i in range(n):
      e_to_2 += x**i/math.factorial(i)
  return e_to_2

x, n = input().split()
x=int(x)
n=int(n)

print(exp(x, n))
```
