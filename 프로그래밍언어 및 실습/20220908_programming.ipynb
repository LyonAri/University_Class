### 기본 자료형

```python
import sys

a=10
b=3.141592
c='a'
d="hello world"
e=True
f=5+1j     #복소수
print(type(a), sys.getsizeof(a))
print(type(b), sys.getsizeof(b))
print(type(c), sys.getsizeof(c))
print(type(d), sys.getsizeof(d))
print(type(e), sys.getsizeof(e))
print(type(f), sys.getsizeof(f))
```

### 산술 연산자

```python
print(10+5)
print(10-5)
print(10*5)
print(10/2)
print(10//3)
print(10%3)
print(10**2)
```

### 비교 연산자

```python
a=10
b=5
print(a < b)
print(a > b)
print(a >= a)
print(a <= a)
print(a == a)
print(a != a)
```

### 논리 연산자

```python
a=True
b=False
print(a and a)
print(a and b)
print(a or b)
print(b or b)
print(not a)
```

```python
a=1.0
b=0.0
print(a and a)
print(a and b)
print(a or b)
print(b or b)
print(not a)
```

### print() 함수

```python
print("Hello World!", end ="\t")
print("Hello", end ="\t")
print("World", end ="\t")
print("Hello\nWorld", end ="\t")
```

```python
name = "홍길동"
age = 23
print("%s의 나이는 %d입니다." % (name, age))
print("{}의 나이는 {}입니다.".format(name, age))
print(f"{name}의 나이는 {age}입니다.")
```

### input() 함수

```python
a = input()  #기본형은 문자열임
b = input()
print(a + b)
```

```python
a = input()
b = input()
print(int(a)+int(b))
```

```python
a = int(input("변수 a의 값을 입력하세요 : "))
b = int(input("변수 b의 값을 입력하세요 : "))
print("a + b의 값은", a + b, "입니다.")
print("a + b의 값은 %d 입니다." % (a + b))
print("a + b의 값은 {} 입니다.".format(a + b))
print(f"a + b의 값은 {a + b} 입니다.")
```

### type() 함수

```python
a = 10
b = 3.141592
c = "Hello World"
d = True
print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(5))
print(type(3.0))
print(type("Python"))
print(type(False))
```
