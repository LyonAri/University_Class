```python
'''
복소수(complex)를 입력받아서 입력받은 수의 거리(r)를 계산하는 프로그램을 작성해보자.
복소수 변수를 z라고 할 경우,
실수 부분과 허수 부분은 z.real 및 z.imag로 접근 가능하다.
복소수의 거리(r)은 소수점 둘째자리까지 출력하도록 한다.
'''
z=input("Complex number : ")

c=complex(z)

x=c.real
y=c.imag

d= (x**2+y**2)**0.5

print("Distance : %.2f" % d)
```

```python
a=int(input("Number a : "))
b=int(input("Number b : "))
c=int(input("Number c : "))

px = (-b+(b**2-(4*a*c))**0.5) / (2*a)
mx = (-b-(b**2-(4*a*c))**0.5) / (2*a)

print("Solution : {}, {}".format(px, mx))
```

```python
age = int(input("나이가 어떻게 되세요? "))
if age >=20:
  print("성인입니다.")
  print("음주가 가능한 나이입니다.")

age = int(input("나이가 어떻게 되세요? "))
if age >=20:
  print("성인입니다.")
  print("음주가 가능한 나이입니다.")
```

```python
h, w = input().split()
h=float(h)
w=float(w)

h_m = h * 0.01
bmi = w / (h_m * h_m)

if bmi <= 18.5:
  print("저체중")
elif bmi <= 23.0:
  print("정상")
elif bmi <= 25.0:
  print("과체중")
elif bmi <= 30.0:
  print("비만1단계")
elif bmi <= 35.0:
  print("비만2단계")
else:
  print("비만 3단계")
```

```python
n=int(input())

if n % 2 == 0:
  print("짝수이며,", end = " ")
else:
  print("홀수이며,", end = " ")
if n % 3 == 0:
  print("3의 배수입니다.")
else:
  print("3의 배수가 아닙니다")
```
